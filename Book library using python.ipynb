```python
# Create class library

class library:
    
    def __init__(self,list,libname):
        self.list=list
        self.libname=libname

# Create 3 function: Display books, LendBook, ReturnBook

    def displaybooks(self):
        print('Books available in the library are: ')
        
        for book in self.list:
            print(book)
            
    def lendbook(self):
        lend=input('Which book do you want: ')
        if lend in self.list:
            self.list.remove(lend)
            print("Book issued")
        else:
            print("Book unavailable")
            
            
    def returnbook(self):
        a=input('Enter the book to return: ')
        self.list.append(a)
        print('Book has been successfully returned')
            
            
if __name__ == '__main__':
    ameya=library(['python', 'java', 'ruby'], 'MyLibrary')
    
    print(f'Welcome to {ameya.libname}')
    print('Services available are: ')
    print('\n')
    print('1. Display books')
    print('2. Lend a book')
    print('3. Return a book')
    print('4. Exit library')
    print('\n')
        

# Prompt the user to select the service

    while (True):                      
        
        choice=int(input('Select the service number you want to access: '))
        
        choice_list=[1,2,3,4]
        
        if choice not in choice_list:
            print('\n')
            print("Choose a valid option")
            continue
        
        elif choice==1:
            print('\n')
            ameya.displaybooks()
            continue
            
        elif choice==2:
            print('\n')
            ameya.lendbook()
            continue
    
        
        elif choice==3:
            print('\n')
            ameya.returnbook()
            continue
            
        else:
            choice==4
            print('\n')
            print("You have exit the library")
        break




            
    
```
